```python
import matplotlib.pyplot as plt
import numpy as np

new=plt.imread('image.jpg')

red=new[:,:,0]
green=new[:,:,1]
blue=new[:,:,2]
gray=0.2989 * red + 0.5870 * green + 0.1140 * blue 
plt.imshow(gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```
